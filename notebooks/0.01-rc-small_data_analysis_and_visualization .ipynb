### Small Data Visualization

```python
# Creating a list for melting dataframe
list_to_melt = [
    'Edu_attain_less_than_High_school_graduate_Population_25_years_and_over',
    'Edu_attain_High_school_graduate_Population_25_years_and_over_',
    'Edu_attain_some_college_or_associates_Population_25_years_and_over',
    'Edu_attain_Bachelor_25_years_and_over_',
    'Edu_attain_graduate_or_professional_25_years_and_over_']

# Creating a melted dataframe to use for plotting
df_melt = df_edu_attain.melt(id_vars='name', value_vars=list_to_melt, var_name='Educational_Attainment', value_name='Population_over_25')

# Renaming Educational Attainment feature for clearity
df_melt.replace({'Edu_attain_less_than_High_school_graduate_Population_25_years_and_over': 'Less Than High School Graduate'}, inplace=True)
df_melt.replace({'Edu_attain_High_school_graduate_Population_25_years_and_over_': 'High School Graduate or Acheived GED'}, inplace=True)
df_melt.replace({'Edu_attain_some_college_or_associates_Population_25_years_and_over': 'Some College or Associates'},inplace=True)
df_melt.replace({"Edu_attain_Bachelor_25_years_and_over_": "Bachelor's Degree"}, inplace=True)
df_melt.replace({'Edu_attain_graduate_or_professional_25_years_and_over_': 'Graduate or Professional Degree'}, inplace=True)
```

```python
# Obtaining list to sort
list_to_sort = list_to_melt

# Creating a bar chart of the different educational attainment distributions
alt.Chart(df_melt, title='Population Statistics On The Level of Educational Attainment in The Southern Boston Area').mark_bar().encode(
    x=alt.X('name', title='Neighborhood'),
    y=alt.Y('Population_over_25', title= 'Population over 25'),
    color=alt.Color('Educational_Attainment:N', sort=list_to_sort),
    tooltip=['name','Educational_Attainment', 'Population_over_25'],
    
).properties(width=700, height=600)
```
